## 1. A simple template for plotting

```python
# step 0: import the necessary packages
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
# step 1: prepare data to plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
# step 2: set up the plot style
mpl.rc('lines', linewidth=4, linestyle='-.')
# step 3: define the layout
fig, ax = plt.subplots()
# step 4: plot the data
ax.plot(x, y, label='sin(x)')
# step 5: add labels and title
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
ax.set_title('sin(x) vs x')
# step 6: add legend
ax.legend()
# step 7: save the plot
fig.savefig('sinx.png')
# step 8: show the plot
plt.show()
```
